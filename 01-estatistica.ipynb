#### RESUMO

```python
import pandas as pd
import numpy as np
```

```python
dict_medidas = {
  'idade': [15, 18, 25, 25, 40, 55, 58, 60, 80],
  'altura': [160, 162, 165, 168, 172, 174, 174, 174, 176],
}
```

```python
df_medidas = pd.DataFrame(dict_medidas)
df_medidas
```

```python
df_medidas.describe()
```

#### MEDIDAS DE POSIÇÃO

```python
# MÉDIA

media = df_medidas['idade'].mean()
media
```

```python
# MEDIANA

mediana = df_medidas['idade'].median()
mediana
```

```python
# MODA
moda = df_medidas['idade'].mode()
moda
```

#### MEDIDAS DE DISPERSÃO

```python
# VARIÂNCIA
variancia = df_medidas['idade'].var()
variancia
```

```python
# DESVIO PADRÃO
desvio_padrao = df_medidas['idade'].std()
desvio_padrao
```

```python
# COEFICIENTE DE VARIAÇÃO
coeficiente_variacao = desvio_padrao / media 
coeficiente_variacao
```

#### MEDIDAS DE FORMA

```python
# ASSIMETRIAS
assimetria_idade = df_medidas['idade'].skew()
assimetria_idade

# Assimetria é positiva.
# A distribuição dos dados é assimétrica à direita, ou seja, a cauda da distribuição se estende mais para a direita do que para a esquerda. 
# Isso sugere que há mais valores altos (maiores) do que valores baixos (menores) na distribuição dos dados.
```

```python
assimetria_altura = df_medidas['altura'].skew()
assimetria_altura

# Assimetria é negativa.
# A distribuição dos dados é assimétrica à esquerda, ou seja, a cauda da distribuição se estende mais para a esquerda do que para a direita. 
# Isso sugere que há mais valores baixos (menores) do que valores altos (maiores) na distribuição dos dados.
```

```python
# CURTOSE
curtose_idade = df_medidas['idade'].kurt()
curtose_idade

# Curtose é positiva.
# A distribuição dos dados é leptocúrtica, ou seja, tem uma cauda mais pesada do que a distribuição normal.
# Isso sugere que há mais valores extremos (outliers) do que o esperado em uma distribuição normal.
```

```python
curtose_altura = df_medidas['altura'].kurt()
curtose_altura  

# Curtose é negativa.
# A distribuição dos dados é platicúrtica, ou seja, tem uma cauda mais leve do que a distribuição normal.
# Isso sugere que há menos valores extremos (outliers) do que o esperado em uma distribuição normal.
```

#### CORRELAÇÃO ENTRE AS VARIÁVEIS

```python
# Correlação do DataFrame segundo o método de Pearson
df_medidas.corr(method='pearson')

# A correlação entre idade e altura é de aproximadamente 0.93, o que indica uma forte correlação positiva entre as duas variáveis.
```

```python
# Correlação entre duas variaveis específicas segundo o método de Pearson
df_medidas['idade'].corr(df_medidas['altura'])
```

```python
# Correlação do DataFrame segundo o método de Spearman
df_medidas.corr(method='spearman')

# A correlação entre idade e altura é de aproximadamente 0.97, o que indica uma forte correlação positiva entre as duas variáveis.
```

##### CRIAR DATAFRAME DE VENDAS

```python
dict_vendas = {
  'categoria': ['masculino', 'feminino', 'infantil', 'casa'],
  'valor': [400000, 600000, 250000, 580000],
  'quantidade': [3000, 5000, 1500, 2500]
}
```

```python
df_vendas = pd.DataFrame(dict_vendas)
df_vendas
```

```python
# GRÁFICO DE BARRAS VERTICAL 
df_vendas.plot.bar(x='categoria', y='valor')
```

```python
# GRÁFICO DE BARRAS VERTICAL (com ordenação)
df_vendas.sort_values('valor', ascending=False).plot.bar(x='categoria', y='valor')
```

```python
# GRÁFICO DE BARRAS HORIZONTAL
df_vendas.sort_values('valor').plot.barh(x='categoria', y='valor')
```

```python
## GRÁFICO DE DISPERSÃO
df_medidas.plot.scatter(x='idade', y='altura')

# Há uma forte correlação positiva entre idade e altura, o que significa que, em geral, à medida que a idade aumenta, a altura também tende a aumentar.
```

```python
# GRÁFICO BOXPLOT
df_medidas.plot.box(column='idade')
```

```python
# GRÁFICO BOXPLOT
df_medidas.plot.box(column='altura')
```
