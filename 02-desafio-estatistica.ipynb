```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dict_faturamento = {
  'data_ref': [
    '2023-01-01', 
    '2020-02-01', 
    '2021-03-01', 
    '2022-04-01', 
    '2023-05-01',
    '2023-06-01', 
    '2020-07-01', 
    '2021-08-01', 
    '2022-09-01', 
    '2023-10-01',
    '2022-11-01', 
    '2023-12-01',
  ],
  'valor': [
    400000, 
    890000, 
    760000, 
    430000, 
    920000,
    340000, 
    800000, 
    500000, 
    200000, 
    900000,
    570000, 
    995000,
  ]
}
```

```python
df_vendas = pd.DataFrame(dict_faturamento)
```

```python
# 1) Média de vendas

media_vendas = df_vendas['valor'].mean()
print(f'A média de vendas é: {media_vendas:.2f}')
```

```python
# 2) Gráfico de vendas ao longo do tempo

df_vendas.sort_values('data_ref').plot.bar(x='data_ref', y='valor')
```

```python
# 3) Gráfico de linhas
df_vendas.sort_values('data_ref').plot.line(x='data_ref', y='valor')
```
